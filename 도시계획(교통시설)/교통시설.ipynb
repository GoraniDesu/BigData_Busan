- 일부 열들의 값들이 누락된 것을 볼 수 있음

```python
# 주차장과 관련된 행만 추출
gdf = gdf[gdf['DGM_NM'].str.contains("주차", na=False)]

# 시군구 코드가 26으로 시작하는 행만 추출 (부산의 시군구 코드 26으로 시작)
gdf = gdf[gdf["SIGNGU_SE"].str.startswith("26")]

gdf.head(5)
```

```python
# 좌표계 확인
print(gdf.crs.to_string())
if gdf.crs.to_string() == 'EPSG:4326':
    print("좌표계가 이미 EPSG:4326입니다.")
else:
    gdf=gdf.to_crs(epsg=4326)
    print("좌표계를 EPSG:4326로 변환하였습니다.")
```

```python
import folium
import folium.features # GeoJson을 사용하기 위해 import

# folium 지도 생성
# 데이터의 평균 좌표를 사용하여 지도의 초기 위치 설정
# 폴리곤의 중심 좌표를 사용하여 초기 위치 설정
m = folium.Map(location=[gdf.geometry.centroid.y.mean(), gdf.geometry.centroid.x.mean()], zoom_start=12)

# 각 폴리곤을 GeoJson 형태로 지도에 추가하고 툴팁 설정
for _, row in gdf.iterrows():
    # 폴리곤 추가 (GeoJson 형태로 변환)
    # 폴리곤에 툴팁 추가
    folium.features.GeoJson(
        row.geometry.__geo_interface__, # shapely geometry를 geojson dictionary로 변환
        name=row['DGM_NM'], # 레이어 컨트롤에 표시될 이름 (선택 사항)
        tooltip=row['DGM_NM'] # 마우스를 올렸을 때 표시될 툴팁
    ).add_to(m)

# 레이어 컨트롤 추가 (선택 사항: 폴리곤/마커 레이어를 켜고 끌 수 있게 함)
folium.LayerControl().add_to(m)

# 지도 저장
m.save('map.html')
```

```python
# gdf를 CSV 파일로 저장합니다.
gdf.to_csv("도시계획(교통시설).csv", index=False, encoding="utf-8")
```

# csv --> shp(공간정보) 형식으로 변환하는 방법

```python
import pandas as pd
from shapely.wkt import loads # CSV에 geometry가 WKT 문자열로 저장되어 있다고 가정

# CSV 파일을 pandas DataFrame으로 읽어옵니다.
df = pd.read_csv("도시계획(교통시설).csv", encoding="utf-8")

# 'geometry' 컬럼의 문자열 데이터를 shapely geometry 객체로 변환합니다.
geometry = df['geometry'].apply(lambda geom_str: loads(geom_str))

# pandas DataFrame과 geometry 컬럼을 사용하여 GeoDataFrame을 생성합니다.
gdf = gpd.GeoDataFrame(df, geometry=geometry, crs="EPSG:4326")

# 공간정보 형식(Shapefile)으로 저장합니다.
gdf.to_file("도시계획(교통시설).shp", driver='ESRI Shapefile', encoding="utf-8")
```
